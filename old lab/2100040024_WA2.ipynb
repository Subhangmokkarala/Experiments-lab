```python
import numpy as np
from scipy import signal
from scipy import misc
import matplotlib.pyplot as plt

image = misc.face(gray=True)
plt.figure()
plt.imshow(image)
plt.gray()
plt.title('Original')

```

```python
noise=np.random.normal(0,150,image.shape)
sine_noise=image+noise
plt.figure()
plt.imshow(sine_noise)
plt.gray()
plt.title('Noise Image')
 
```

```python
sine_noise1 = image+noise*0.05
plt.figure()
plt.imshow(sine_noise1)
plt.gray()
plt.title('clean Image')
```
